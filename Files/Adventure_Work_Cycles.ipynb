```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Q1: What is the relationship between Country and Revenue?


```python
df = pd.read_csv("Q1 country vs revenue.csv")
print(df)
```

```python
plt.scatter('Country', "Number of Stores", s='Annual Revenue', alpha=0.5, data=df)
plt.ylim([0, 700])
plt.xlim([-0.5,6])
plt.xlabel('Country')
plt.ylabel('Number of Stores')
plt.title("Country Annual Revenue and Number of Stores")
plt.xticks(rotation=90)
plt.text(1, -300, 'Note:  Annual Revenue (in 100,000)')
plt.show()
```

### Q3: What is the relationship between store trading duration and revenue?

```python
df_a = pd.read_csv('Q3_results.csv')

fig, axes = plt.subplots()
plt.subplots_adjust(wspace=0.3, hspace=0.4)

x_a = df_a['YearOpened']
y_a = df_a['avg_rev']
m_a, b_a = np.polyfit(x_a, y_a, 1)
axes.scatter(x_a, y_a, color='orange', alpha=0.7)
axes.plot(x_a, m_a*x_a + b_a, color='blue', linewidth=2)
axes.set_xlabel('Year Opened')
axes.set_ylabel('Average Revenue')
axes.set_title('Trading Duration and Revenue')
axes.grid(True)
```

### Q4: What is the relationship between the size of the stores, number of employees and revenue?

### Q5: What is the relationship between annual leave taken and bonus?

```python
df = pd.read_csv('Q5_results.csv')
print(df)
plt.figure(figsize=(8, 6))
plt.scatter(df['OrganizationLevel'], df['Bonus'], df['VacationHours'],color='red', alpha=0.5)
plt.xticks(np.arange(-1, 6, step=1))
#plt.plot(df['VacationHours'], df['Bonus'], marker='h',ms=15, mfc='g', mec='r', ls='none')
plt.xlabel('Organization Level')
plt.ylabel('Bonus')
plt.title('Relationship between Organization Level, Annual Leave Taken and Bonus')
plt.text(-2, -2000, "Note: 0 is for CEO, 1 is for executive management, 2 for middle management, 3 for supervisory management, 4 for non-management")
plt.grid(True)
plt.show()
```
